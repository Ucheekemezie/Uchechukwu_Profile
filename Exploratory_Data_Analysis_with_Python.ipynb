# Exploratory Data Analysis on a Lemonade Sales Dataset using Python

## 1. Introduction
In this project, we conduct Exploratory Data Analysis (EDA) on a sample lemonade sales dataset using Python. The dataset includes information about sales volume, daily temperature, and weather conditions (sunny or not). The aim is to uncover insights and identify patterns through descriptive statistics and visualizations, simulating a typical data analyst’s workflow when examining new data.



## 2. Importing Required Libraries
We begin by importing the necessary Python libraries for data analysis and visualization:
- **pandas** is used for data manipulation and analysis.
- **matplotlib.pyplot** is used for creating static, interactive, and animated visualizations.
- **seaborn** is built on top of Matplotlib and simplifies the creation of informative visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 3. Creating a Sample Dataset
The dataset consists of 7 entries, each representing a day of the week. It was created manually using a Python dictionary and converted to a Pandas DataFrame.
The dataset includes the following columns:
•	Day of the week
•	Temperature (in Celsius)
•	Weather condition (Sunny: Yes/No)
•	Sales (in cups of lemonade)


```python
data = {
  'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
  'Temperature': [20, 25, 22, 30, 28, 32, 18],
  'Sunny': ['No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'No'],
  'Sales': [30, 50, 35, 70, 60, 80, 20]
  }
df = pd.DataFrame(data)
print(df)
```

## 4. Inspecting the Dataset
Using the ‘df.info()’ and ‘df.describe()’ methods shown below, we confirmed that:
•	There are no missing values in the dataset.
•	All variables are of appropriate data types.
•	The average temperature is 25°C, and the average sales volume is around 49 cups.
•	Temperature and sales show some variability, as indicated by standard deviation values.

```python
print(df.info())
```

```python
print(df.head())
```

```python
print(df.describe())
```

## 5. Calculating the Descriptive Statistics
Key statistics are computed as shown below.These metrics suggest a balanced dataset with moderate variation in both weather and customer behavior across the week.

```python
# Mean, median, and standard deviation for Temperature and Sales
mean_temp = df['Temperature'].mean()
median_temp = df['Temperature'].median()
std_temp = df['Temperature'].std()
mean_sales = df['Sales'].mean()
median_sales = df['Sales'].median()
std_sales = df['Sales'].std()
print(f"Temperature - Mean: {mean_temp:.2f}, Median: {median_temp:.2f}, Std Dev: {std_temp:.2f}")
print(f"Sales - Mean: {mean_sales:.2f}, Median: {median_sales:.2f}, Std Dev: {std_sales:.2f}")
```

## 6. Visualizing Data - Histogram
The histogram of lemonade sales shows a fairly symmetrical distribution. Sales are relatively evenly spread across the week, with no strong skew. This suggests sales are fairly consistent and influenced more by other factors such as weather or temperature.


```python
# Create a histogram to explore the distribution of lemonade sales.

plt.figure(figsize=(8, 5))
sns.histplot(df['Sales'], bins=5, kde=True)
plt.title('Distribution of Lemonade Sales')
plt.xlabel('Sales (Cups)')
plt.ylabel('Frequency')
plt.show()
```

## 7. Visualizing Data - Scatter Plot
The scatter plot of Sales vs. Temperature demonstrates a clear positive relationship, particularly on sunny days. Sales tend to increase with rising temperature, and the visual separation by 'Sunny' labels highlights the potential influence of weather on customer behaviour.


```python
# Create a scatter plot to investigate the relationship between temperature and sales.

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Temperature', y='Sales', hue='Sunny', data=df)
plt.title('Sales vs. Temperature')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Sales (Cups)')
plt.show()
```

## 8. Visualizing Data - Bar Chart
A bar chart comparing average sales by weather condition shows significantly higher sales on sunny days than on non-sunny ones. This supports the hypothesis that sunshine boosts lemonade demand. This finding can be used for future sales predictions or inventory planning.


```python
# Compare average sales on sunny vs. non-sunny days using a bar chart.

plt.figure(figsize=(8, 5))
sns.barplot(x='Sunny', y='Sales', data=df)
plt.title('Average Sales by Weather Condition')
plt.xlabel('Sunny (Yes/No)')
plt.ylabel('Average Sales (Cups)')
plt.show()
```

## 9. Correlation Analysis
Using Pearson correlation, we found a perfect positive correlation of 1.00 between Temperature and Sales. This indicates that as temperature increases, lemonade sales increase proportionally. While this is insightful, the small sample size means we should be cautious about generalizing the result without more data.

```python
# Calculate the correlation between Temperature and Sales to quantify their relationship.

correlation = df['Temperature'].corr(df['Sales'])
print(f"Correlation between Temperature and Sales: {correlation:.2f}")
```
